## METHODS

```python
def generate_random_data(start, num_imgs, img_size, min_object_size, max_object_size, path, labels):

  for i in range(start, num_imgs):
    # Generate the name
    name = str(i) + ".jpeg"

    # Generate the image
    img = np.zeros((img_size, img_size, 1), dtype = "uint8")
    img.fill(255)

    # Generate the position of the block
    w, h = np.random.randint(min_object_size, max_object_size, size=2)
    x = np.random.randint(0, img_size - w)
    y = np.random.randint(0, img_size - h)

    tl_x = x
    tl_y = y + h
    br_x = x + w
    br_y = y
    cv2.rectangle(img, (tl_x, tl_y), (br_x, br_y), (0), -1)

    # Save the generated image. 
    cv2.imwrite(os.path.join(path , name), img)

    # Add the detection box data to the list
    detection_box = [tl_x, tl_y, br_x, br_y]
    labels.append(detection_box)
  return labels
```

```python
def generate_seq_data(num_imgs, img_size, min_object_size, max_object_size, path):
  labels = []
  
  w, h = np.random.randint(min_object_size, max_object_size, size=2)
  w = 20
  h = 20

  x = 0
  y = 0

  for i in range(num_imgs):
    # Generate the name
    name = str(i) + ".jpeg"

    # Generate the image
    img = np.zeros((img_size, img_size, 1), dtype = "uint8")
    img.fill(255)

    # Generate the position of the block
    # x = np.random.randint(0, img_size - w)
    # y = np.random.randint(0, img_size - h)

    x += 1
    y += 1

    tl_x = x
    tl_y = y + h
    br_x = x + w
    br_y = y
    cv2.rectangle(img, (tl_x, tl_y), (br_x, br_y), (0), -1)

    # Save the generated image. 
    cv2.imwrite(os.path.join(path , name), img)

    # Add the detection box data to the list
    detection_box = [tl_x, tl_y, br_x, br_y]
    labels.append(detection_box)

  pd.DataFrame(labels).to_csv(os.path.join(path , 'detection_boxes.csv' + ".csv"))
```
